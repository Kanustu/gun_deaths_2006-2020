```python
import pandas as pd
```

```python
def load_csv(csv):
    return pd.read_csv(csv)
```

```python
combined_df = pd.DataFrame()
for x in range(2006, 2021):
   current_df = pd.read_csv(f'../gun_deaths_2006-2020/gun_data_2006-2020/guns_{x}.csv')
   combined_df = pd.concat([combined_df, current_df], ignore_index=True)
```

```python
combined_df.drop('hispanic', axis=1, inplace=True)
```

```python
combined_df.rename(columns={'sex':'victim_sex', 'age':'victim_age', 'race':'victim_race', 'education':'victim_education'}, inplace=True)
```
